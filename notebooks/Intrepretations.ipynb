## Top 20 Feature Importances

```python
# Extracting feature importances
feature_importances = rf_best.feature_importances_

# Creating DataFrame for feature importances
importances_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})

# Sorting the DataFrame based on importance
importances_df = importances_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.barplot(x="Importance", y="Feature", data=importances_df.head(20), palette="viridis")

plt.title('Top 20 Feature Importances in RandomForestRegressor', fontsize=16)
plt.xlabel('Importance', fontsize=14)
plt.ylabel('Feature', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

plt.show()
```

### Top 10 Feature Importances

```python
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.barplot(x="Importance", y="Feature", data=importances_df.head(10), palette="viridis")

plt.title('Top 10 Feature Importances in RandomForestRegressor', fontsize=16)
plt.xlabel('Importance', fontsize=14)
plt.ylabel('Feature', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

plt.show()
```

#### Comments: 

Features like previous_365_days_log_avg_likes, previous_365_days_log_avg_views, and title_length_chars have the highest importance scores, indicating they are highly influential in the model's decision-making process.

Interpreting the Features:

1. Features related to historical data (previous_365_days_...) suggest that past performance metrics of a video (or a channel's video) such as likes, views, and comment counts are strong predictors for future view counts.
2. Thumbnail-related features (Saturation, Brightness, Contrast) indicate that the visual appeal of a video's thumbnail can significantly influence its view counts.
3. Title-related features (title_length_chars, title_avg_word_length, title_all_lowercase) imply that certain characteristics of a video's title, like its length and use of capitalization, play a role in attracting viewers.

**This indicates that both content quality (as perceived through engagement metrics) and presentation (metadata and thumbnail) are key factors in predicting the popularity of YouTube videos.**

```python
X_train = pd.read_csv("X_train.csv")

X_train
```

```python
# Initialize the SHAP explainer with your model
explainer = shap.TreeExplainer(rf_best)

# Calculate SHAP values for the dataset you provided
shap_values = explainer.shap_values(X_train)

# Plot the SHAP values - we'll use a simple summary plot here
shap.summary_plot(shap_values, X_train, plot_type="bar")
```

#### Intrepretations:

The x-axis label "mean(|SHAP value|) (average impact on model output magnitude)" indicates that the values are the average of the absolute SHAP values for each feature, a measure of each feature's overall importance without considering the direction of the effect. 

1. The features such as previous_365_days_log_avg_likes, previous_365_days_log_avg_views, and previous_365_days_geo_avg_like_dislike_ratio seem to be key drivers in the model's decisions._

2. The color of the bar represents the direction of the relationship between the feature and the target variable. It seems that all bars are of the same color, which typically indicates a positive correlation: higher values of these features tend to increase the prediction output (views) of the model.

Comparing the results from the top 20 feature importances plot and the SHAP summary plot, the RF model emphasizes the importance of features like brightness, saturation, contrast, and video category. From both plots, the key features of significance are previous likes, log average views, and the length of the title. Interestingly, the like-dislike ratio feature also shows up in the SHAP plot, as well as more features on the title like exclaimation count and case. The RF top feature plot suggests that thumbnail features like higher brightness, saturation, and contrast levels may lead to increased views, while the choice of video category can significantly impact viewership. It also highlights the relevance of the video's timing, both in the local and normalized context. However, the SHAP summary plot reveals that the effects of these features on views are not strictly linear. Furthermore, the SHAP values for features like category and publishing time indicate that their impact on views varies across different videos. This highlights the need for a holistic understanding of content creation strategies, as factors influencing viewership are multifaceted and context-dependent.

## Model Performances

```python
data = {
    'Model': ['RandomForestRegressor', 'XGBRegressor', 'Lasso', 'Baseline 1', 'Baseline 2'],
    'MAE': [2081482.96, 2071499.53, 3239764.38, 3324987.69, 2192965.96],
    'MSE': [56436427243572.86, 64559375522716.67, 140310015863701.23, 147921560488945.4, 57477095865023.94],
    'R2': [0.614950479, 0.559529938, 0.042705093, -0.009226288, 0.607850296]
}

df_performance = pd.DataFrame(data)

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(10, 5))

# MAE Plot
sns.barplot(y='MAE', x='Model', data=df_performance, ax=axes[0], palette='viridis')
axes[0].set_title('Model Comparison by MAE')
axes[0].tick_params(axis='x', rotation=45)

# MSE Plot
sns.barplot(y='MSE', x='Model', data=df_performance, ax=axes[1], palette='viridis')
axes[1].set_title('Model Comparison by MSE')
axes[1].tick_params(axis='x', rotation=45)

# R2 Plot
sns.barplot(y='R2', x='Model', data=df_performance, ax=axes[2], palette='viridis')
axes[2].set_title('Model Comparison by R2')
axes[2].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()

plt.show()
```

#### Intrepretation:

Among the three regression models evaluated, the RF model stands out with the lowest Mean Absolute Error (MAE) and a relatively R2 value of 0.615. This suggests that the RF model provides the most accurate predictions among the models, with its predictions, on average, being closest to the actual values, and it explains approximately 61.5% of the variance in the data. 

The XGBoost model, while still performing reasonably well, exhibits slightly higher errors and a slightly lower R2 value compared to the RandomForestRegressor. On the other hand, the Lasso model lags significantly behind, showing both high MAE and a very low R2 value, indicating poor predictive performance. 

In the baseline models, particularly Baseline 2 (predicting views based on previous video views), outperform Baseline 1 (predicting average views), emphasizing the importance of incorporating more relevant features for improved predictions.
